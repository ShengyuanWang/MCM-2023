```python
# Define a function that takes a 5-letter word as input
def classify_word_structure(word):
    vowels = 'aeiou'
    consonants = 'bcdfghjklmnpqrstvwxyz'

    # Check the pattern of vowels and consonants in the word
    if word[0] in consonants and word[1] in vowels and word[2] in consonants and word[3] in vowels and word[4] in consonants:
        return 1 # "VCVCV" # e.g. lolly, candy
    elif word[0] in consonants and word[1] in vowels and word[2] in consonants and word[3] in consonants and word[4] in vowels:
        return 2 # "VCCCV" # e.g. crisp, blend
    elif word[0] in vowels and word[1] in consonants and word[2] in vowels and word[3] in consonants and word[4] in vowels:
        return 3 # "VCVCV" # e.g. oasis, abaci
    elif word[0] in vowels and word[1] in consonants and word[2] in consonants and word[3] in vowels and word[4] in consonants:
        return 4 # "VCVCV" # e.g. music, lemon
    else:
        return 0 # "Unknown" # The word does not match any known pattern

word_vc_structure = []
for word in data['Word']:
    word_vc_structure.append(classify_word_structure(word))
data['word_vc_structure'] = word_vc_structure
data[['Word', 'word_vc_structure']]
```

```python
# 一个单词中重复字母出现的次数
# Define a function that takes a word as input
def classify_word_by_repeated_letters(word):
    # Convert the word to lowercase
    word = word.lower()

    # Create a dictionary to count the frequency of each letter in the word
    letter_counts = {}
    for letter in word:
        if letter in letter_counts:
            letter_counts[letter] += 1
        else:
            letter_counts[letter] = 1

    # Find the maximum frequency of any letter in the word
    max_frequency = max(letter_counts.values())

    # Classify the word based on the maximum frequency of any letter
    if max_frequency == 1:
        return 1 #"No repeated letters" # e.g. hello, world
    elif max_frequency == 2:
        return 2 #"Two letters repeated" # e.g. candy, llama
    elif max_frequency == 3:
        return 3 #"Three letters repeated" # e.g. balloon, assess
    else:
        return 0#"More than three letters repeated" # e.g. bookkeeper, Mississippi

max_repeat = []
for word in data['Word']:
    max_repeat.append(classify_word_by_repeated_letters(word))
data['max_repeat'] = max_repeat
data[['Word', 'max_repeat']]
```

```python
# rate familarity method 1

import nltk
from nltk.corpus import brown

# Define a function that takes a word as input
def rate_word_by_frequency(word):
    # Convert the word to lowercase
    word = word.lower()

    # Get the frequency of the word in the Brown Corpus
    word_freq = brown.words().count(word)

    # Get the total number of words in the Brown Corpus
    total_words = len(brown.words())

    # Calculate the frequency of the word as a percentage of the total words
    freq_percentage = (word_freq / total_words) * 100

    # Return the frequency percentage
    return freq_percentage
```

```python
# rate familarity method 2

import requests

# Define a function that takes a word as input and returns its frequency of occurrence in the Google Books Ngram Viewer corpus
def get_word_frequency(word):
    # Construct the URL to query the corpus
    url = "https://books.google.com/ngrams/json?content=" + word + "&year_start=1800&year_end=2019&corpus=26&smoothing=3"

    # Send a request to the URL and get the response
    response = requests.get(url)

    # Parse the response as JSON and extract the frequency of occurrence of the word
    if response.status_code == 200:
        data = response.json()
        if len(data) > 0 and len(data[0]['timeseries']) > 0:
            return sum(data[0]['timeseries']) # the sum of the frequencies across all years
    return 0 # the word was not found in the corpus

# Define a function that rates a word based on the frequency of occurrence in the corpus
def rate_word_familiarity(word):
    frequency = get_word_frequency(word)*100000
    return frequency

familiarity = []
for word in data['Word']:
    familiarity.append(rate_word_familiarity(word))
```
