```python
# packages
import pandas as pd
```

```python
data = pd.read_csv('data/q1_cleand.csv')
data['percentage'] = data['Hard']/data['Reported']
data
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np
scaler = MinMaxScaler()
percentage_scalar = np.array(data['percentage']).reshape(-1, 1)
data['percentage_scalar'] = scaler.fit_transform(percentage_scalar)
vowel_rate_scalar = np.array(data['vowel_rate']).reshape(-1, 1)
data['vowel_rate_scalar'] = scaler.fit_transform(vowel_rate_scalar)
max_repeat_scalar = np.array(data['max_repeat']).reshape(-1, 1)
data['max_repeat_scalar'] = scaler.fit_transform(max_repeat_scalar)
```

```python
corr_matrix = data[['percentage', 'vowel_rate','word_vc_structure','max_repeat', 'familiarity']].corr()
```

```python
import seaborn as sns
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
```
